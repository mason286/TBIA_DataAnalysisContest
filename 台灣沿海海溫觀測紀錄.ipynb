```python
import json
import pandas as pd

# 讀取 JSON 文件
file_path = "taiwan_sea_temperature.json"
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 獲取資源數據
resource_data = data['cwaopendata']['Resources']['Resource']['Data']['SeaSurfaceObs']['Location']

# 展平數據並提取溫度數據
records = []
for location in resource_data:
    location_info = {
        'StationID': location['Station']['StationID'],
        'StationName': location['Station']['StationName'],
        'StationLatitude': location['Station']['StationLatitude'],
        'StationLongitude': location['Station']['StationLongitude'],
    }
    monthly_data = location['StationObsStatistics']['Monthly']
    for month_data in monthly_data:
        if 'Temperature' in month_data:
            record = location_info.copy()
            record['Time'] = month_data['Date']
            record['SeaTemperature'] = month_data['Temperature']
            records.append(record)

# 創建 DataFrame
df = pd.DataFrame(records)

# 將時間列轉換為日期格式
df['Time'] = pd.to_datetime(df['Time'])

# 選擇2005年到2023年之間的數據
df_selected = df[(df['Time'].dt.year >= 2005) & (df['Time'].dt.year <= 2023)]

# 計算平均海溫
average_sea_temperature = df_selected['SeaTemperature'].mean()

print(f"2005-2023年的平均海溫為: {average_sea_temperature:.2f} °C")
```

```python
import json
import pandas as pd

# 讀取 JSON 文件
file_path = "taiwan_sea_temperature.json"
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 獲取資源數據
resource_data = data['cwaopendata']['Resources']['Resource']['Data']['SeaSurfaceObs']['Location']

# 展平數據並提取溫度數據
records = []
for location in resource_data:
    location_info = {
        'StationID': location['Station']['StationID'],
        'StationName': location['Station']['StationName'],
        'StationLatitude': location['Station']['StationLatitude'],
        'StationLongitude': location['Station']['StationLongitude'],
    }
    monthly_data = location['StationObsStatistics']['Monthly']
    for month_data in monthly_data:
        data_month = month_data['DataMonth']  # 月份
        sea_temperature = month_data['SeaTemperature']  # 海溫數據
        
        # 安全檢查是否能夠正確轉換為浮點數
        try:
            max_temp = float(sea_temperature['Maximum'])
            min_temp = float(sea_temperature['Minimum'])
            mean_temp = float(sea_temperature['Mean'])
        except ValueError:
            continue  # 跳過無效數據
        
        record = location_info.copy()
        record['Month'] = int(data_month)  # 轉換月份為整數
        record['MaxSeaTemperature'] = max_temp  # 最大海溫
        record['MinSeaTemperature'] = min_temp  # 最小海溫
        record['MeanSeaTemperature'] = mean_temp  # 平均海溫
        records.append(record)

# 創建 DataFrame
df = pd.DataFrame(records)

# 打印前幾行數據來檢查
df.head()

```

```python
import json
import pandas as pd

# 讀取 JSON 文件
file_path = "taiwan_sea_temperature.json"
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 獲取資源數據
resource_data = data['cwaopendata']['Resources']['Resource']['Data']['SeaSurfaceObs']['Location']

# 定義目標站點名稱
target_station_name = '台中潮位站'

# 收集目標站點的每月平均海溫數據
monthly_means = []

# 遍歷資源數據，查找萡子寮潮位站的資料並提取每月平均海溫
for location in resource_data:
    if location['Station']['StationName'] == target_station_name:
        monthly_data = location['StationObsStatistics']['Monthly']
        for month_data in monthly_data:
            # 提取日期和平均海溫
            data_month = int(month_data['DataMonth'])  # 月份
            mean_sea_temp = float(month_data['SeaTemperature']['Mean'])  # 平均海溫
            year = int(month_data['SeaTemperature']['MaximumYear'])  # 年份
            monthly_means.append({
                'Year': year,
                'Month': data_month,
                'MeanSeaTemperature': mean_sea_temp
            })
            
# 創建 DataFrame
df = pd.DataFrame(monthly_means)

# 選擇2005年到2023年之間的數據
df_selected = df[(df['Year'] >= 2005) & (df['Year'] <= 2023)]

# 打印結果
print(f"塭港潮位站 2005年到2023年每個月平均海溫:")
print(df_selected[['Year', 'Month', 'MeanSeaTemperature']])
```

```python
import json
import pandas as pd

# 讀取 JSON 文件
file_path = "taiwan_sea_temperature.json"
with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 獲取資源數據
resource_data = data['cwaopendata']['Resources']['Resource']['Data']['SeaSurfaceObs']['Location']

# 定義目標站點名稱
target_station_name = '臺中港潮位站'

# 收集目標站點的每月平均海溫數據
monthly_means = []

# 遍歷資源數據，查找萡子寮潮位站的資料並提取每月平均海溫
for location in resource_data:
    if location['Station']['StationName'] == target_station_name:
        monthly_data = location['StationObsStatistics']['Monthly']
        for month_data in monthly_data:
            # 提取日期和平均海溫
            data_month = int(month_data['DataMonth'])  # 月份
            mean_sea_temp = float(month_data['SeaTemperature']['Mean'])  # 平均海溫
            year = int(month_data['SeaTemperature']['MaximumYear'])  # 年份
            monthly_means.append({
                'Year': year,
                'Month': data_month,
                'MeanSeaTemperature': mean_sea_temp
            })
            
# 創建 DataFrame
df = pd.DataFrame(monthly_means)

# 選擇2005年到2023年之間的數據
df_selected = df[(df['Year'] >= 2005) & (df['Year'] <= 2023)]

# 按照年份和月份排序
df_sorted = df_selected.sort_values(by=['Year', 'Month'])

# 重新設置索引，並打印結果
df_sorted = df_sorted.reset_index(drop=True)
print(df_sorted[['Year', 'Month', 'MeanSeaTemperature']])

excel_file_path = "臺中港潮位站sea_temperature_data.xlsx"
df_sorted.to_excel(excel_file_path, index=False)

print(f"Excel 檔案已成功生成: {excel_file_path}")
```
